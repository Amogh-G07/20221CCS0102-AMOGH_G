```python
import folium

# List of locations (latitude, longitude)
locations = [[1.366623,103.821285], [1.283873,103.845202], [1.348382,103.852376]]

# Create a map centered at the first location
mymap = folium.Map(location = locations[0],
                   width=950,
                   height=550,
                   zoom_start=12,
                   tiles='openstreetmap')

# Add markers for each location
for location in locations:
    folium.Marker(location=location).add_to(mymap)

# Display the map
mymap









import folium

# List of locations (latitude, longitude)
locations = [[1.366623,103.821285], [1.283873,103.845202], [1.348382,103.852376]]

# List of popup texts
popups = ['Marker 1', 'Marker 2', 'Marker 3']

# Create a map centered at the first location
mymap = folium.Map(location = locations[0],
                   width=950,
                   height=550,
                   zoom_start=12,
                   tiles='openstreetmap')

# Add markers with popups for each location
for location, popup in zip(locations, popups):
    folium.Marker(location=location, popup=popup).add_to(mymap)

# Display the map
mymap
```

```python
import folium

# Create a map centered at Bangalore location
mymap = folium.Map(location =[12.9716,77.5946],
                   width=950,
                   height=550,
                   zoom_start=12,
                   tiles='openstreetmap')

# Add a marker to the map
folium.Marker(location=[12.9716,77.5946], popup='Bangalore City').add_to(mymap)

# Add an image overlay to the map
folium.raster_layers.ImageOverlay(
    image='https://upload.wikimedia.org/wikipedia/commons/thumb/1/11/Vidhana_Soudha%2C_14_February_2022_-_cropped.jpg/300px-Vidhana_Soudha%2C_14_February_2022_-_cropped.jpg',
    bounds=[[12.9650, 77.5750], [12.9780, 77.6150]],
    opacity=0.5,
    interactive=True
).add_to(mymap)

# Display the map
mymap
```
