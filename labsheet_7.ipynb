```python
import pandas as pd
import matplotlib.pyplot as plt

data={
    "Dates":pd.to_datetime(["2022-01-01","2022-01-01","2022-02-02","2022-03-03","2022-04-04","2022-05-05","2022-06-06",]),
    "Value":[10,15,22,28,35,42,48]
}

df=pd.DataFrame(data)
df.set_index("Dates",inplace=True)

plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker='X')
plt.xlabel("Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01"]),
    "value":[10,20,17,24,16,18,15,20,17]
    }
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='x')
plt.xlabel("Date")
plt.ylabel("values")
plt.title("Time series with seasonality")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

text_data="Python is a popular programming language"
wordcloud=WordCloud(width=800,height=400,background_color='black').generate(text_data)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.title("word cloud")
plt.axis('off')
plt.show()


words=text_data.split()
word_counts=Counter(words)
plt.figure(figsize=(10,5))
plt.bar(word_counts.keys(),word_count.values())
plt.title('Histogram of word frequencies')
plt.xlabel('words')
plt.ylabel('frequency')
plt.xticks(rotation=90)
plt.show()
```
